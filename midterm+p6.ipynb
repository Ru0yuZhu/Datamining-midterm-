```python
import pandas as pd
import numpy as np
data = pd.read_excel('C:\school work\data mining\midterm\\final_dataset_spring_18.xls')
data.head()
```

```python
df = data.iloc[:,1:39]
df.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
%matplotlib inline

```

```python
df.replace(' ', np.nan, inplace=True)
df
```

```python
null_number = sum(df.isnull().sum())
null_number
```

```python
df_new = df.fillna(df.mean())
df_new
```

```python
from scipy import cluster
```

```python
initial = [cluster.vq.kmeans(df_new, i) for i in range(1,12)]
plt.plot([var for (cent,var) in initial])
plt.show()
```

```python
import itertools
mat = df_new.as_matrix()
merged = list(itertools.chain.from_iterable(mat))
type(merged)
a = pd.DataFrame(merged)
a
```

```python
km = KMeans(n_clusters=4)
km.fit(a)
# Get cluster assignment labels
labels = km.labels_
```

```python
results = pd.DataFrame([a.index,labels]).T
results
```

```python
g = a.join(results.iloc[:,1])
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(g.iloc[:,1], g.iloc[:,0])
plt.show()
```
